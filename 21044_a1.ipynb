# Problem 1


```python
import math
import matplotlib.pyplot as plt
import numpy as np

TRUE_ROOT = 0.56714329

def fx1(x):
    return math.exp(-1*x)-x

def dfx1(x):
    return -1*math.exp(-1*x)-1

x = np.linspace(0, 1,10)

plt.plot(x, [fx1(a) for a in x] )
plt.show()
```

```python
def bisection(xl, xu, n):
    error = []
    for i in range(n):
        xr = (xl+xu)/2
        error.append(abs((TRUE_ROOT-xr)/TRUE_ROOT))

        if fx1(xl)*fx1(xr)<0:
            xu = xr
        else:
            xl = xr

    return error, xr
    
def false_position(xl, xu, n):
    error = []
    for i in range(n):
        xr = (xu - fx1(xu)*(xl - xu)/(fx1(xl) - fx1(xu)))
        error.append(abs((TRUE_ROOT-xr)/TRUE_ROOT))

        if fx1(xl)*fx1(xr)<0:
            xu = xr
        else:
            xl = xr
    
    return error, xr

def newton_raphson(x, n):
    error = []
    for i in range(n):
        x = x -fx1(x)/dfx1(x)
        error.append(abs((TRUE_ROOT-x)/TRUE_ROOT))

    return error, x


```

```python
n=20

bi_error, bi_root = bisection(0.4,0.5, n)
false_pos_error, false_pos_root = false_position(0.4,0.5, n)
newton_raph_error, newton_raph_root = newton_raphson(0.4, n)

plt.yscale('log')
plt.plot(np.arange(n), bi_error, c='r')
plt.plot(np.arange(n), false_pos_error, c='b')
plt.plot(np.arange(n), newton_raph_error, c='g')
plt.show()

print('Root from bisection:', bi_root)
print('Root from false positon:', false_pos_root)
print('Root from newton raphson:', newton_raph_root)

```

# Problem 2

```python
def f2x(x):
    return pow(x,20)-1


x = np.linspace(0.9, 1.1,20)

plt.plot(x, [f2x(a) for a in x] )
plt.show()
```

```python
def df2x(x):
    return 20*pow(x,19)

def bisection(xl, xu, n):
    error = []
    for i in range(n):
        xr = (xl+xu)/2
        error.append(abs((1-xr)/1))

        if f2x(xl)*f2x(xr)<0:
            xu = xr
        else:
            xl = xr

    return error, xr
    
def false_position(xl, xu, n):
    error = []
    for i in range(n):
        xr = (xu - f2x(xu)*(xl - xu)/(f2x(xl) - f2x(xu)))
        error.append(abs((1-xr)/1))

        if f2x(xl)*f2x(xr)<0:
            xu = xr
        else:
            xl = xr
    
    return error, xr

def newton_raphson(x, n):
    error = []
    for i in range(n):
        x = x -f2x(x)/df2x(x)
        error.append(abs((1-x)/1))

    return error, x
```

```python
n=20

bi_error, bi_root = bisection(0.9,1.1, n)
false_pos_error, false_pos_root = false_position(0.9,1.1, n)
newton_raph_error, newton_raph_root = newton_raphson(1.1, n)

plt.yscale('log')
plt.plot(np.arange(n), bi_error, c='r')
plt.plot(np.arange(n), false_pos_error, c='b')
plt.plot(np.arange(n), newton_raph_error, c='g')
plt.show()

print('Root from bisection:', bi_root)
print('Root from false positon:', false_pos_root)
print('Root from newton raphson:', newton_raph_root)
```

```python
def false_position(xl, xu, n):
    for i in range(n):
        xr = (xu - f3x(xu)*(xl - xu)/(f3x(xl) - f3x(xu)))

        if f3x(xl)*f3x(xr)<0:
            xu = xr
        else:
            xl = xr
    
    return xr

root = false_position(0.07, 0.08, 20)
print(f'Root is {root} with value {f3x(root)}')

```
